```python
''' Beam parameters '''
N = 4096			# number of sampling points
E0 = 9481			# photon energy [eV]
tau = 400			# pulse duration [fs]
window = 10000		# total simulation time window [fs]

''' OE parameters '''
#hkl = [6,6,0]
hkl = [4,4,0]
#hkl = [3,3,3] 

# parameter dictionary. z_source is in LCLS coordinates (20 meters upstream of undulator exit)
beam_params = {
	'photonEnergy': E0,
	'N': N,
	'sigma_x': 23e-6,
	'sigma_y': 23e-6,
	'rangeFactor': 5,
	'scaleFactor': 10,
	'z_source': 0
}
```

```python
def define_devices(slit_width = 1e-3, E0=18e3, FOV=2e-3):

    # viewing point upstream of premono
    im0 = optics.PPM('im0', z=157.05, FOV=3e-3, N=256)
    
    # premono
    c1 = optics.Crystal('c1_pre', hkl=[2,2,0], length=10e-2, width=20e-3, z=279.05, E0=E0,
                              alphaAsym=0.296706, orientation=3, pol='s', delta=0.e-6)
    c2 = optics.Crystal('c2_pre', hkl=[4,4,0], length=10e-2, width=20e-3, z=279.09, E0=E0,
                              alphaAsym=0., orientation=1, pol='s', delta=0.e-6)
    c3 = optics.Crystal('c3_pre', hkl=[4,4,0], length=10e-2, width=20e-3, z=279.34, E0=E0,
                              alphaAsym=0., orientation=3, pol='s', delta=0.e-6)
    c4 = optics.Crystal('c4_pre', hkl=[2,2,0], length=10e-2, width=20e-3, z=279.55, E0=E0,
                              alphaAsym=-0.296706, orientation=1, pol='s', delta=0.e-6)
    # viewing point after premono
    im1 = optics.PPM('after_C4', z=279.551, FOV=3e-3, N=256)

    # first crystal: symmetric reflection
    crystal1 = optics.Crystal('c1', hkl=hkl, length=10e-2, width=20e-3, z=289.47, E0=E0,
                              alphaAsym=0, orientation=3, pol='s', delta=0.e-6)

    # second crystal: asymmetric reflection, orientation flipped relative to crystal1
    crystal2 = optics.Crystal('c2', hkl=hkl, length=10e-2, width=20e-3, z=289.55, E0=E0,alphaAsym=-0.51487212, 
                              orientation=1,pol='s', delta=0e-6)
    
    # printing crystal incidence and reflection angles for confirmation
    print('crystal 2 incidence angle: {:.2f} degrees'.format(crystal2.alpha*180/np.pi))
    print('crystal 2 exit angle: {:.2f} degrees'.format(crystal2.beta0*180/np.pi))

    im2 = optics.PPM('after_HRM2', z=299.549, FOV=2e-3, N=256)

    # CRL with ~1 meter focal length (modeled as single element for now)
    crl1 = optics.CRL('crl1', z=299.55, E0=E0, f=26.3773, diameter=1.5e-3)

    # viewing point downstream of 1st crl
    im3 = optics.PPM('after_F1', z=29.551,N=256,FOV=2e-3)

    # slit at focus
    slit = optics.Slit('slit', z=309.55, x_width=slit_width, y_width=2e-3)

    # viewing point at focus
    im4 = optics.PPM('after_WDS', z=309.551, FOV=1.5e-3, N=256)

    # second CRL with ~1 meter focal length, for collimation
    crl2 = optics.CRL('crl2', z=319.55, E0=E0, f=26.3773, diameter=1.5e-3)
    
    # viewing point downstream of 2nd crl
    im5 = optics.PPM('after_F2', z=319.551,N=256,FOV=1e-3)
    
    # viewing point before 3rd crystal
    im6 = optics.PPM('before_HRM3', z=329.549,N=256,FOV=1e-3)

    # third crystal, symmetric reflection, same orientation as crystal2
    crystal3 = optics.Crystal('c3', hkl=hkl, length=10e-2, width=10e-3, z=329.55, E0=E0,alphaAsym=0.51487212,
                              orientation=1, asym_type='emergence',pol='s')
    
    im7 = optics.PPM('after_HRM3', z=329.551,N=256,FOV=1e-3)
    
    # fourth crystal, asymmetric reflection, same orientation as crystal1
    crystal4 = optics.Crystal('c4', hkl=hkl, length=10e-2, width=10e-3, z=crystal3.z + (crl1.z-crystal2.z)*np.cos(crystal1.beta0*2), E0=E0,alphaAsym=-alphaAsym, 
                              asym_type='emergence', orientation=3,pol='s')

    # viewing point just downstream of monochromator
    im8 = optics.PPM('after_HRM4', z=crystal4.z+1e-3, FOV=5e-3, N=256)
    
    im9 = optics.PPM('after_HRM', z=crystal4.z+10., FOV=5e-3, N=256)

    # list of devices to propagate through
    devices = [im0, c1,c2,c3,c4,im1, crystal1,crystal2,im2, crl1,im3,slit,im4,crl2,im5, im6,crystal3,im7,crystal4,im8, im9]

    return devices
```

```python
''' Define beamline '''
try: alphaAsym
except:
    crystal_temp = optics.Crystal('crystal', hkl=hkl, length=10e-2, width=20e-3, z=930, E0=E0,
                                  alphaAsym=0, orientation=0, pol='s', delta=0.e-6)

    alphaAsym = crystal_temp.alpha - np.deg2rad(5)      # calculate miscut angle (5 degree grazing incidence)

    if alphaAsym <= 0:
        print('\n***\n***\n*** Bragg angle smaller than grazing angle')

devices = define_devices(slit_width = 200e-6, E0=E0)
```

```python
%%capture capt
''' propagate '''
mono_beamline = beamline.Beamline(devices)

tstart = time.time()
beam_params['photonEnergy'] = E0
# pulse -> beam
pulse = beam.Pulse(beam_params=beam_params, tau=tau, time_window=window)
screen_names=['im0','after_C4','after_HRM2','after_F1','after_WDS','after_F2','before_HRM3','after_HRM3','after_HRM4','after_HRM']
pulse.propagate(beamline=mono_beamline, screen_names=screen_names)
tfin = time.time()
```

```python
print('{} images'.format(len(screen_names)))
print('propagation time: {}s'.format(np.round(tfin-tstart,1)))
```

```python
''' spatial projection (x,y)'''
for i in range(len(screen_names)):
    pulse.imshow_projection(screen_names[i])
```

```python
''' plot time slices '''
print('total {}s, per slice {}ms'.format(round(tfin-tstart,2), round(1000*(tfin-tstart)/pulse.N,2)))

shift = 0
for i in range(len(screen_names)):
    pulse.imshow_time_slice(screen_names[i],shift=shift)
```
