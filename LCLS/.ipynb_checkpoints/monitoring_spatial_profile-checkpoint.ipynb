```python
''' Beam parameters '''
N = 4096			# number of sampling points
E0 = 9481			# photon energy [eV]
tau = 400			# pulse duration [fs]
window = 10000		# total simulation time window [fs]

''' OE parameters '''
#hkl = [6,6,0]
hkl = [4,4,0]
#hkl = [3,3,3] 

# parameter dictionary. z_source is in LCLS coordinates (20 meters upstream of undulator exit)
beam_params = {
	'photonEnergy': E0,
	'N': N,
	'sigma_x': 23e-6,
	'sigma_y': 23e-6,
	'rangeFactor': 5,
	'scaleFactor': 10,
	'z_source': 630
}
```

```python
def define_devices(
    f1, f2, slit_width = 1e-3, hkl = [1,1,1], alphaAsym = 0., E0=18e3, f0 = 290., d23=7.):

    # viewing point upstream of monochromator
    im0 = optics.PPM('im0', z=870, FOV=5e-3, N=256)
    crl0 = optics.CRL('crl0', z=920, E0=E0, f=f0, diameter=2e-3)

    # first crystal: symmetric reflection
    crystal1 = optics.Crystal('c1', hkl=hkl, length=10e-2, width=20e-3, z=930, E0=E0,
                              alphaAsym=0, orientation=0, pol='s', delta=0.e-6)
    im1 = optics.PPM('After C1', z=crystal1.z+.01, FOV=5e-3, N=256)

    # second crystal: asymmetric reflection, orientation flipped relative to crystal1
    crystal2 = optics.Crystal('c2', hkl=hkl, length=10e-2, width=20e-3, z=crystal1.z+.2, E0=E0,alphaAsym=alphaAsym, 
                              orientation=2,pol='s', delta=0e-6)
    im2 = optics.PPM('After C2', z=crystal2.z+.01, FOV=5e-3, N=256)
    
    # printing crystal incidence and reflection angles for confirmation
    print('crystal 2 incidence angle: {:.2f} degrees'.format(crystal2.alpha*180/np.pi))
    print('crystal 2 exit angle: {:.2f} degrees'.format(crystal2.beta0*180/np.pi))


    # CRL with ~1 meter focal length (modeled as single element for now)
    crl1 = optics.CRL('crl1', z=crystal2.z+f1, E0=E0, f=f2, diameter=5e-3)
    im3 = optics.PPM('After CRL1', z=crl1.z+.01,N=256,FOV=5e-3)

    # slit at focus
    slit = optics.Slit('slit', z=crl1.z+f2, x_width=slit_width, y_width=2e-3)
    im4 = optics.PPM('Before Slit', z=slit.z-.01, FOV=5e-3, N=256)
    im5 = optics.PPM('After Slit', z=slit.z+.01, FOV=5e-3, N=256)

    # second CRL with ~1 meter focal length, for collimation
    crl2 = optics.CRL('crl2', z=crl1.z+2*f2, E0=E0, f=f2, diameter=5e-3)
    im6 = optics.PPM('After CRL2', z=crl2.z+.01,N=256,FOV=5e-3)
    
    

    # third crystal, symmetric reflection, same orientation as crystal2
    crystal3 = optics.Crystal('c3', hkl=hkl, length=10e-2, width=10e-3, z=crl2.z+d23, E0=E0,alphaAsym=0, orientation=2,
                             asym_type='emergence',pol='s')
    im7 = optics.PPM('Before C3', z=crystal3.z-.01,N=256,FOV=5e-3)
    im8 = optics.PPM('After C3', z=crystal3.z+.01,N=256,FOV=5e-3)
    
    # fourth crystal, asymmetric reflection, same orientation as crystal1
    crystal4 = optics.Crystal('c4', hkl=hkl, length=10e-2, width=10e-3, z=crl2.z+d23 + (f1-d23)*np.cos(crystal1.beta0*2), E0=E0,alphaAsym=-alphaAsym, 
                              asym_type='emergence', orientation=0,pol='s')
    im9 = optics.PPM('After C4', z=crystal4.z+.01,N=256,FOV=5e-3)
    
    im10 = optics.PPM('output', z=crystal4.z+10.,N=256,FOV=5e-3)

    # list of devices to propagate through
    devices = [im0,crl0,
               crystal1,im1,    crystal2,im2,    crl1,im3,      im4,slit,im5,
               crl2,im6,         im7,crystal3,im8,       crystal4,im9,
               im10]
    return devices

screen_names=['im0',
              'After C1',      'After C2',      'After CRL1',   'Before Slit','After Slit',
              'After CRL2',     'Before C3','After C3',    'After C4',
              'output']
```

```python
''' Define beamline '''
try: alphaAsym
except:
    crystal_temp = optics.Crystal('crystal', hkl=hkl, length=10e-2, width=20e-3, z=930, E0=E0,
                                  alphaAsym=0, orientation=0, pol='s', delta=0.e-6)

    alphaAsym = crystal_temp.alpha - np.deg2rad(5)      # calculate miscut angle (5 degree grazing incidence)

    if alphaAsym <= 0:
        print('\n***\n***\n*** Bragg angle smaller than grazing angle')

devices = define_devices(10., 10., slit_width = 200e-6, hkl=hkl, alphaAsym=alphaAsym, E0=E0)
```

```python
print('asymmetry angle: {}'.format(alphaAsym))
```

```python
%%capture capt
''' propagate '''
mono_beamline = beamline.Beamline(devices)

tstart = time.time()
beam_params['photonEnergy'] = E0
# pulse -> beam
pulse = beam.Pulse(beam_params=beam_params, tau=tau, time_window=window)
pulse.propagate(beamline=mono_beamline, screen_names=screen_names)
tfin = time.time()
```

```python
print('{} images'.format(len(screen_names)))
print('propagation time: {}s'.format(np.round(tfin-tstart,1)))
```

```python
''' spatial projection (x,y)'''
for i in range(len(screen_names)):
    pulse.imshow_projection(screen_names[i])
```

```python
''' plot time slices '''
print('total {}s, per slice {}ms'.format(round(tfin-tstart,2), round(1000*(tfin-tstart)/pulse.N,2)))

shift = 0
for i in range(len(screen_names)):
    pulse.imshow_time_slice(screen_names[i],shift=shift)
```
